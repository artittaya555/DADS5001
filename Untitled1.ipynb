```python
import pandas as pd
import plotly.graph_objects as go

# 1) อ่านและเตรียมข้อมูล
url = "https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv"
df = pd.read_csv(url, sep="\t")

# แปลงราคาเป็นตัวเลข (ตัด $ ออก)
df['item_price'] = df['item_price'].str.replace('$', '', regex=False).astype(float)

# 2) หา min / max ต่อ item_name
g = (
    df.groupby('item_name', as_index=False)['item_price']
      .agg(min='min', max='max')
)

# เรียงให้อ่านง่าย (เอาตาม max จากน้อยไปมาก)
g = g.sort_values('max', ascending=True).reset_index(drop=True)

# (ถ้ารายการเยอะมาก อยากจำกัด Top N ก็ใช้ g = g.tail(40) อะไรแบบนี้ได้)

# 3) เตรียมเส้นนอนต่อแถว (min -> max -> None)
line_x, line_y = [], []
for _, row in g.iterrows():
    line_x.extend([row['min'], row['max'], None])
    line_y.extend([row['item_name'], row['item_name'], None])

# 4) วาดกราฟ
fig = go.Figure(
    data=[
        # เส้นนอนเชื่อม min ถึง max
        go.Scatter(
            x=line_x,
            y=line_y,
            mode="lines",
            showlegend=False,
            line=dict(color="grey")
        ),
        # จุด min (ซ้าย)
        go.Scatter(
            x=g['min'],
            y=g['item_name'],
            mode="markers",
            name="min",
            marker=dict(color="green", size=9),
            hovertemplate="Item: %{y}<br>Min: $%{x:.2f}<extra></extra>"
        ),
        # จุด max (ขวา)
        go.Scatter(
            x=g['max'],
            y=g['item_name'],
            mode="markers",
            name="max",
            marker=dict(color="blue", size=9),
            hovertemplate="Item: %{y}<br>Max: $%{x:.2f}<extra></extra>"
        ),
    ]
)

fig.update_layout(
    title="Chipotle: Min vs Max item_price by item_name",
    xaxis_title="item_price (USD)",
    yaxis_title="item_name",
    xaxis_tickprefix="$",
    height=max(500, 20 * len(g))  # ปรับความสูงตามจำนวนรายการ
)

fig.show()
```
